### 1. Importing libraries and dataset

```python
# Importing libraries and scaler
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import os
import sklearn
from sklearn.preprocessing import StandardScaler
```

```python
# Defining a path for importing/exporting
path = r'/Users/peterguan/ClimateWins ML'
```

```python
# Importing the ClimateWins dataset
climatewins = pd.read_csv(r'/Users/peterguan/ClimateWins ML/Data Sets/Dataset-weather-prediction-dataset-processed.csv', index_col = False)
```

```python
# Checking the first 5 rows
climatewins.head()
```

```python
# Checking the dimensions of the dataset
climatewins.shape
```

```python
# Resetting index and Renaming column
climatewins.reset_index(inplace = True)
climatewins.rename(columns={'index':'id'}, inplace = True)
```

```python
# Checking if the column was renamed successfully
climatewins.head()
```

```python
climatewins.info()
```

### 2. Cleaning the dataset

```python
# Checking the dimensions of the dataset
climatewins.shape
```

```python
# Checking for missing values
climatewins.isnull().sum()
```

```python
# Dropping all the missing values
climatewins.dropna(inplace = True)
```

```python
# Checking if any rows were dropped
climatewins.shape
```

```python
# Checking for duplicates
climatewins_duplicates = climatewins[climatewins.duplicated]
```

```python
# No duplicates found
climatewins_duplicates
```

### 3. Scaling the dataset

```python
# Creating a scaler object using StandardScaler from sklearn.preprocessing
scaler = StandardScaler()
```

```python
# Creating a new dataframe with the Scaled data. (ADDED "columns=climatewins.columns" so the column names transferred over)

climatewins_scaled = pd.DataFrame(scaler.fit_transform(climatewins), columns=climatewins.columns)
climatewins_scaled.head()
```

```python
# Comparing the old dataframe with the new Scaled dataframe
climatewins.head()
```

```python
# Replacing the 'date' column in the Scaled dataframe with the 'date' column from the original dataframe
climatewins_scaled['DATE'] = climatewins['DATE']
```

```python
# Checking if the replacement was successful
climatewins_scaled.head()
```

### 4. Exporting the dataset

```python
# Exporting dataframe as a .csv
climatewins_scaled.to_csv(os.path.join(path, 'climatewins_scaled.csv'))
```
